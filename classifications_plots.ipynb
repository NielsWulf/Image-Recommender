```python
import pickle
from collections import Counter
import matplotlib.pyplot as plt

# Load classification data
with open("classifications_with_names.pkl", "rb") as f:
    classifications = pickle.load(f)

# Extract class names
class_names = [info["class_name"] for info in classifications.values()]

# Count the frequency of each class name
class_counts = Counter(class_names)

# Get the top 20 most common classes
top_classes = class_counts.most_common(50)

# Separate class names and their counts for plotting
class_labels, counts = zip(*top_classes)

# Plotting the data using matplotlib
plt.figure(figsize=(12, 8))
plt.barh(class_labels, counts, color="skyblue")
plt.xlabel("Number of Images")
plt.ylabel("Class Name")
plt.title("Top 20 Most Frequent Image Classes")
plt.gca().invert_yaxis()  # Invert y-axis to have the most frequent class on top
plt.show()
```

```python
import pickle
from collections import Counter
import plotly.express as px

# Load classification data
with open("classifications_with_names.pkl", "rb") as f:
    classifications = pickle.load(f)

# Extract class names
class_names = [info["class_name"] for info in classifications.values()]

# Count the frequency of each class name
class_counts = Counter(class_names)

# Get the top 20 most common classes
top_classes = class_counts.most_common(50)

# Separate class names and their counts for plotting
class_labels, counts = zip(*top_classes)

# Create a bubble chart using plotly
fig = px.scatter(
    x=class_labels,
    y=[1] * len(class_labels),  # Use a constant y-value for all classes
    size=counts,
    color=counts,
    hover_name=class_labels,
    size_max=60,
    labels={"x": "Class Name", "y": ""},
    title="Top 20 Most Frequent Image Classes",
)

# Update layout to improve visualization
fig.update_layout(
    xaxis=dict(title="Class Name", showgrid=False),
    yaxis=dict(showticklabels=False),
    showlegend=False,
)

# Show the figure
fig.show()
```
